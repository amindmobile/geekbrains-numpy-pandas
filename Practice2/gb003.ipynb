<h1>Задание 1</h1>

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

Данные о недвижимости

```python
data = boston.data
```

Target - цель, что нужно предсказать, в этом случае цена на недвижимость

```python
target = boston.target
```

Названия признаков (пока не ясно зачем)

```python
feature_names = boston.feature_names
```

Создаю датафреймы

X - признаки недвижимости

```python
x = pd.DataFrame(data, columns=feature_names)
```

Y - данные о цене на недвижимость (которую надо в итоге предсказать)

```python
y = pd.DataFrame(target, columns=['price'])
```

Разделю данные на датасеты: тренировочный и тестовый

```python
from sklearn.model_selection import train_test_split
```

На выборке "train" обучаю модель. На выборке "test" проверяю наксколько хорошо модель работает, сверяя значение цены с предсказанными ценами на недвижимость.
X - данные признаков, Y - данные о target (в д.с. цене на недвижимость).
test_size - размер рандомной тестовой выборки. random_state - "повторимость рандома test_state" (про повторимость я пока не вкурил).

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

<b>Создаю модель линейной регрессии</b> (офигеть)

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

fit - тренирует модель (почему оно на выводе просто скобки кажет, а в видео есть вывод внутри них?)

```python
lr.fit(x_train, y_train)
```

predict - предсказывает цену на недвижимость в тестовой выборке

```python
y_pred = lr.predict(x_test)
```

Делаю датафрейм с реальными и предсказанными ценами для визуальной оценки точности предсказания (нафиг так, лучше графиком. Сиди сравнивай тут цифры..)

```python
check_test = pd.DataFrame({'y_test': y_test['price'],
                           'y_pred': y_pred.flatten()},
                         columns=['y_test', 'y_pred'])
```

```python
check_test["ошибка"] = check_test["y_pred"] - check_test["y_test"]
check_test.head(10)
```

<i>Метрики оценки качества</i> покурить плотнее (R2 чем ближе к единице тем лучше)

```python
from sklearn.metrics import r2_score
```

```python
r2_score_l=r2_score(check_test["y_pred"], check_test["y_test"])
r2_score_l
```

```python
model_gscv.fit(x_train, y_train)
```
